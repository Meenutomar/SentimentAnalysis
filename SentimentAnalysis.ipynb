```python
# prompt: reviews from https://www.pcmag.com/categories/processors/brands/intel#:~:text=Latest%20Intel%20Processor%20Reviews&text=Intel%27s%20power-hungry%2C%20pricey%20Core,comes%20much%20more%20highly%20recommende in this i have page attribute with page number 1to 3 collect the all data and scrap them

import requests
from bs4 import BeautifulSoup

# Initialize an empty list to store all reviews
all_reviews = []

# Loop through pages 1 to 3
for page_number in range(1, 4):
    # Construct the URL for the current page
    url = f"https://www.pcmag.com/categories/processors/brands/intel#:~:text=Latest%20Intel%20Processor%20Reviews&text=Intel%27s%20power-hungry%2C%20pricey%20Core,comes%20much%20more%20highly%20recommende?page={page_number}"

    # Make a GET request to fetch the raw HTML content
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract all the reviews from the page
    reviews = soup.find_all("div", class_="w-full")

    # Append the reviews to the list
    all_reviews.extend(reviews)

# Print each review
for review in all_reviews:
    print(review.text)
```

```python
# prompt: convert this all_reviews data into dataframe

import pandas as pd

# Create a list of dictionaries where each dictionary represents a review
review_data = []
for review in all_reviews:
  review_data.append({"review": review.text})

# Create a Pandas DataFrame from the list of dictionaries
df = pd.DataFrame(review_data)

# Print the DataFrame
print(df)
```
